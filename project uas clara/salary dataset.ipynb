```python
#Import semua library yang dibutuhkan
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score, mean_squared_error
from scipy.stats import iqr

#Load dataset (Salary Dataset - Simple Linear Regression)
data = pd.DataFrame({
    'YearsExperience': [1.1, 1.3, 1.5, 2.0, 2.2, 2.9, 3.0, 3.2, 3.2, 3.7, 3.9,
                        4.0, 4.0, 4.1, 4.5, 4.9, 5.1, 5.3, 5.9, 6.0, 6.8,
                        7.1, 7.9, 8.2, 8.7, 9.0, 9.5, 9.6, 10.3, 10.5],
    'Salary': [39343, 46205, 37731, 43525, 39891, 56642, 60150, 54445, 64445,
               57189, 63218, 55794, 56957, 57081, 61111, 67938, 66029, 83088,
               81363, 93940, 91738, 98273, 101302, 113812, 109431, 105582,
               116969, 112635, 122391, 121872]
})

#Statistik Deskriptif
mean_salary = data['Salary'].mean()
median_salary = data['Salary'].median()
std_salary = data['Salary'].std()
iqr_salary = iqr(data['Salary'])

#Regresi Linier
X = data[['YearsExperience']]
y = data['Salary']
model = LinearRegression()
model.fit(X, y)
y_pred = model.predict(X)

r2 = r2_score(y, y_pred)
mse = mean_squared_error(y, y_pred)
slope = model.coef_[0]
intercept = model.intercept_

#Ringkasan Analisis
print("Statistik Deskriptif Gaji:")
print(f"Rata-rata     : Rp {mean_salary:,.0f}")
print(f"Median        : Rp {median_salary:,.0f}")
print(f"Standar Deviasi: Rp {std_salary:,.0f}")
print(f"IQR           : Rp {iqr_salary:,.0f}")
print("\n Model Regresi:")
print(f"Persamaan     : Gaji = {slope:.2f} × Tahun + {intercept:,.0f}")
print(f"R² Score      : {r2:.3f}")
print(f"MSE           : {mse:,.0f}")
```

```python
#Visualisasi 1 : Scatter plot + Garis regresi
plt.figure(figsize=(10,6))
sns.regplot(x='YearsExperience',y='Salary',
data=data,ci=None,color='skyblue',line_kws={"color":"red"})
plt.title('Visualisasi  1 : Regresi Linier antara pengalaman dan Gaji',fontsize=14)
plt.xlabel('Tahun Pengalaman')
plt.ylabel('Gaji(Rp)')
plt.grid(True)
plt.tight_layout()
plt.show()
          
```
