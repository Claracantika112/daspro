<b>Project Data Peserta Audisi Menyanyi </b> <br>
Buatlah contoh data peserta audisi menyanyi sebanayak 1000 yang akan digunkan untuk prediksi:<br>
<ol>
    <li> Usia:	15 - 30
    <li>Gender:	L/P
    <li>Kota Asal:	Jakarta, Bandung, Surabaya, Medan, Makassar
    <li>Jenis Musik: Pop, Rock, Jazz, Dangdut, K-Pop
    <li>Kategori Audisi: Solo, Duo, Grup
    <li>Pelatihan:	Ya, Tidak
    <li>Nilai Vokal: 1 - 10
    <li>Nilai Penampilan: 1 - 10
    <li>Nilai Total: Rata-rata dari vokal dan penampilan
    <li>Status Lolos: Lolos, Tidak Lolos (berdasarkan nilai total â‰¥ 7.0)



```python
import numpy as np
import pandas as pd

# Jumlah data
n_data = 1000

# Generate data peserta audisi
usia = np.random.randint(15, 31, size=n_data)
gender = np.random.choice(['L', 'P'], size=n_data)
kota_asal = np.random.choice(['Jakarta', 'Bandung', 'Surabaya', 'Medan', 'Makassar'], size=n_data)
jenis_musik = np.random.choice(['Pop', 'Rock', 'Jazz', 'Dangdut', 'K-Pop'], size=n_data)
kategori = np.random.choice(['Solo', 'Duo', 'Grup'], size=n_data)
pelatihan = np.random.choice(['Ya', 'Tidak'], size=n_data)
nilai_vokal = np.random.randint(1, 11, size=n_data)
nilai_penampilan = np.random.randint(1, 11, size=n_data)
nilai_total = (nilai_vokal + nilai_penampilan) / 2
status_lolos = np.where(nilai_total >= 7, 'Lolos', 'Tidak Lolos')

# Buat DataFrame
df = pd.DataFrame({
    'Usia': usia,
    'Gender': gender,
    'Kota Asal': kota_asal,
    'Jenis Musik': jenis_musik,
    'Kategori Audisi': kategori,
    'Pelatihan': pelatihan,
    'Nilai Vokal': nilai_vokal,
    'Nilai Penampilan': nilai_penampilan,
    'Nilai Total': nilai_total,
    'Status Lolos': status_lolos
})

# Tampilkan info dasar
print(df.info())
print()

# Statistik dasar
print(f"Usia termuda peserta : {df['Usia'].min()}")
print(f"Usia tertua peserta  : {df['Usia'].max()}")
print()

# Jumlah berdasarkan gender
print(f"Jumlah peserta laki-laki : {len(df[df['Gender'] == 'L'])}")
print(f"Jumlah peserta perempuan : {len(df[df['Gender'] == 'P'])}")
print()

# Jumlah berdasarkan kota
for kota in df['Kota Asal'].unique():
    print(f"Jumlah peserta dari {kota:<10}: {len(df[df['Kota Asal'] == kota])}")
print()

# Jumlah berdasarkan jenis musik
for genre in df['Jenis Musik'].unique():
    print(f"Jumlah peserta genre musik {genre:<7}: {len(df[df['Jenis Musik'] == genre])}")
print()

# Jumlah berdasarkan kategori audisi
for kat in df['Kategori Audisi'].unique():
    print(f"Jumlah peserta kategori {kat:<5}: {len(df[df['Kategori Audisi'] == kat])}")
print()

# Pelatihan
print(f"Jumlah peserta yang ikut pelatihan : {len(df[df['Pelatihan'] == 'Ya'])}")
print(f"Jumlah peserta yang tidak ikut     : {len(df[df['Pelatihan'] == 'Tidak'])}")
print()

# Nilai rata-rata
print(f"Rata-rata nilai vokal      : {df['Nilai Vokal'].mean():.2f}")
print(f"Rata-rata nilai penampilan : {df['Nilai Penampilan'].mean():.2f}")
print(f"Rata-rata nilai total      : {df['Nilai Total'].mean():.2f}")
print()

# Status lolos
print(f"Jumlah peserta lolos       : {len(df[df['Status Lolos'] == 'Lolos'])}")
print(f"Jumlah peserta tidak lolos : {len(df[df['Status Lolos'] == 'Tidak Lolos'])}")
print()

# Simpan ke CSV
df.to_csv('dataPesertaAudisi.csv', index=False)
print("Data berhasil disimpan ke 'dataPesertaAudisi.csv'")
```
